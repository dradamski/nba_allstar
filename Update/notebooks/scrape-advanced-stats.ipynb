```python
from bs4 import BeautifulSoup, Comment
import numpy as np
import pandas as pd

from data_grabbing import (go_to_player_page,
                           get_df_starts,
                           pick_stats,
                           get_stats_from_page
                          )
```

```python
with open(sys.path[-1]+'/players.txt') as f:
    mylist = [tuple(map(str, i.split(','))) for i in f]

player_list = [(i[0].strip("'('"), i[1].strip(" )\n'")) for i in mylist]
```

```python
len(player_list)
```

```python
%%time
all_df = pd.DataFrame()
for player in player_list:
    player_soup = go_to_player_page(player)
    all_list, df_starts = get_df_starts(player_soup)
    stat_indexes = pick_stats(['per100', 'advanced'], df_starts)
    
    player_df = pd.DataFrame()
    
    for stat_index in stat_indexes:   
        if player_df.empty:
            player_df = get_stats_from_page(all_list, stat_index, player)
            
        else:
            player_df = player_df.merge(get_stats_from_page(all_list, stat_index, player), 
                            on=['Season', 'Age', 'Tm', 
                                'Lg', 'Pos', 'Player',
                                'href'], 
                            suffixes=('_per100','_adv'))
    all_df = all_df.append(player_df)
    print(len(all_df.Player.unique()), str(player_df.Player.unique()), player_df.shape[0])
    
```
